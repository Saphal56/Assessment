```python
# //data2['Longitude'].fillna(data2['Longitude'].mean(), inplace=True)
import re
#re module provides support for regular expressions 

# Assuming the 'my_column' column is the string column to be converted to float
numeric_values = data2['Longitude'].str.extract(r'(\d+\.?\d*)', expand=False)
data2['Longitude'] = numeric_values.astype(float)
numeric_values = data2['Latitude'].str.extract(r'(\d+\.?\d*)', expand=False)
data2['Latitude'] = numeric_values.astype(float)
```
